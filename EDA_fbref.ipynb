```python
# List of dataset paths
file_paths = [
    "Fbref-data/merged_premier_league_2020_2024.csv",
    "Fbref-data/merged_laliga_2020_2024.csv",
    "Fbref-data/merged_bundesliga_2020_2024.csv",
    "Fbref-data/merged_ligue1_2020_2024.csv",
    "Fbref-data/merged_serieA_2020_2024.csv"
]

for file_path in file_paths:
    print(f"\n📂 Checking missing values in: {file_path}")

    try:
        df = pd.read_csv(file_path)

        # Calculate percentage of missing values per column
        missing_ratio = df.isnull().mean() * 100
        missing_ratio = missing_ratio[missing_ratio > 0].sort_values(ascending=False)

        if missing_ratio.empty:
            print("✅ No missing values found.")
        else:
            print(f"❗ Columns with missing values ({len(missing_ratio)}):")
            print(missing_ratio)

    except Exception as e:
        print(f"⚠️ Error processing {file_path}: {e}")
```

```python
# We remove the features with more than 5% of missing values in each data set (5% missing for all data set)
# Columns to drop
columns_to_drop = [
    'Standard G/SoT',
    'Standard G/Sh',
    'Standard SoT%',
    'Standard Dist',
    'Take-Ons Tkld%',
    'Take-Ons Succ%',
    'Challenges Tkl%',
    'Aerial Duels Won%',
    'Long Cmp%', 
    '90s_possession', 
    '90s_defense',
    '90s_misc', 
    '90s_shooting', 
    'Expected npxG/Sh', 
    'Expected_shooting npxG/Sh'
]

# Loop through and process each file
for file_path in file_paths:
    print(f"\n📂 Processing: {file_path}")
    
    try:
        df = pd.read_csv(file_path)

        # Drop columns that exist in this dataset
        drop_cols = [col for col in columns_to_drop if col in df.columns]
        df.drop(columns=drop_cols, inplace=True)

        # Save cleaned DataFrame back to the same file
        df.to_csv(file_path, index=False)
        print(f"✅ Dropped {len(drop_cols)} columns and saved: {file_path}")
    
    except Exception as e:
        print(f"⚠️ Error processing {file_path}: {e}")
```

```python
# replace missing values with median for other variables : 

for file_path in file_paths:
    print(f"\n📂 Processing: {file_path}")

    try:
        df = pd.read_csv(file_path)

        numeric_cols = df.select_dtypes(include='number').columns
        for col in numeric_cols:
            if df[col].isnull().any():
                median_val = df[col].median()
                df[col].fillna(median_val, inplace=True)
                print(f"  ✅ Filled missing values in '{col}' with median: {median_val:.2f}")

        # Save cleaned dataset
        df.to_csv(file_path, index=False)
        print(f"✅ Saved cleaned file: {file_path}")

    except Exception as e:
        print(f"⚠️ Error processing {file_path}: {e}")
```

```python

for file_path in file_paths:
    print(f"\n📂 Final filling for: {file_path}")

    try:
        df = pd.read_csv(file_path)

        # === Clean and fill age columns ===
        age_cols = [col for col in df.columns if col.startswith("age") and df[col].isnull().any()]
        for col in age_cols:
            # Extract the part before the dash, convert to numeric
            df[col] = df[col].astype(str).str.extract(r'^(\d+)')[0]
            df[col] = pd.to_numeric(df[col], errors='coerce')

            if df[col].isnull().any():
                median_age = df[col].median()
                df[col].fillna(median_age, inplace=True)
                print(f"  ✅ Cleaned and filled '{col}' with median: {median_age:.1f}")

        # === Fill nation columns with mode ===
        nation_cols = [col for col in df.columns if col.startswith("nation") and df[col].isnull().any()]
        for col in nation_cols:
            mode_nation = df[col].mode().iloc[0] if not df[col].mode().empty else "Unknown"
            df[col].fillna(mode_nation, inplace=True)
            print(f"  🌍 Filled '{col}' with mode: {mode_nation}")

        # Save cleaned data
        df.to_csv(file_path, index=False)
        print(f"✅ All cleaned and saved: {file_path}")

    except Exception as e:
        print(f"⚠️ Error processing {file_path}: {e}")
```

```python
import os
import shutil

# Exported file prefix (you can change this)
export_dir = "Fbref-data/Cleaned_final_data"
os.makedirs(export_dir, exist_ok=True)

# Export files by copying them to export_dir
for file_path in file_paths:
    try:
        file_name = os.path.basename(file_path)
        export_path = os.path.join(export_dir, file_name)
        shutil.copy(file_path, export_path)
        print(f"✅ Exported: {export_path}")
    except Exception as e:
        print(f"⚠️ Failed to export {file_path}: {e}")
```
